# Proyecto: Estimación de salarios mediante regresión lineal simple

## Modelo: $y = a_{0} + a_{1}x $, donde

## $y$     <-- es el valor a predecir

## $x$     <-- son variables independientes

## $a_{0}$ y $a_{1}$ <-- son los pesos que se encuentran en el entrenamiento

### * Cargamos datos usando pandas

### * Separamos datos para entrenamiento y pruebas mediante `train_test_split`

### * Usaremos la función `score` para evaluar precisión de las predicciones

```python
import pandas as pd 
import matplotlib.pyplot as plt 

# libreria que nos permite dividir los datos para entrenamiento y pruebas:
from sklearn.model_selection import train_test_split

# libreria para regresión lineal:
from sklearn.linear_model import LinearRegression
```

```python
# Cargamos base de datos:
dataset = pd.read_csv('./Datos/salarios.csv')
```

```python
# visualizamos los campos de los primeros 5 registros:
dataset.head(5)
```

```python
dataset.shape
```

```python
# Separamos datos de entrenamiento y prueba:

# la primer columna coresponde a los "features":
x = dataset.iloc[:,0].values # <-- extraemos primer columna
x = x.reshape(x.shape[0],1) # redimensionamos a un vector columna

# la segunda columna coresponde a los "labels":
y = dataset.iloc[:,1].values

# Separamos los datos:
p = 0.8 # <-- porcentaje de datos de entrenamiento
X_train , X_test , Y_train , Y_test = train_test_split(x, y, train_size = 0.8, random_state = 0 )
# random_state =0 <--- indica que cada vez que ejecutamos el modelo, dicho conjunto sea fijo
```

```python
# Cargamos el método de entrenamiento:
model_regression = LinearRegression()

# Entrenamos el modelo:
model_regression.fit(X_train,Y_train)
```

```python
# Aplicamos el aprendizaje a los datos de entrenamiento:
Y_pred_train = model_regression.predict(X_train)

# Aplicamos el aprendizaje a los datos de prueba:
Y_pred_test = model_regression.predict(X_test)
```

```python
# Visualizamoe el entrenamiento:
viz_train = plt 
viz_train.scatter( X_train , Y_train, color = 'blue' ,label = 'data train')
viz_train.plot(X_train , Y_pred_train , color = 'black', label = 'model trained')
viz_train.title('Resultados del entrenamiento')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.legend()
viz_train.show()
```

```python
# Visualizamoe el aprendizaje:
viz_train = plt 
viz_train.scatter( X_test , Y_test, color = 'red', label = 'data test')
viz_train.plot(X_test , Y_pred_test , color = 'black' , label = 'model')
viz_train.title('Resultado de pruebas')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.legend()
viz_train.show()
```

```python
# Aplicamos el model_regression a un dato de prueba en particular:

j=5

Y_perd_j = model_regression.predict( [X_test[j]] ) 

print('Etiqueta real = ', Y_test[j])



print('\nEtiqueta predicha = ', Y_perd_j)
```

```python
# Evaluamos la puntiación del aprendizaje:
c = model_regression.score(X_test , Y_test)

print('El modelo puede predecir con un ',c*100,' % de precisión ')
```
